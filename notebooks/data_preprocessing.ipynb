```python
#Creating day,month and year columns for all the datasets

def extract_date_parts(datasets, date_columns):
    for name, df in datasets.items():
        if name in date_columns:
            for col in date_columns[name]:
                if col in df.columns:
                    df[f"{col}_Month_Number"]= df[col].dt.month
                    df[f'{col}_Month'] = df[col].dt.strftime('%b')
                    df[f"{col}_Year"] = df[col].dt.year
            print(f"Extracted date parts for {name}")
        else:
            print(f"No date column specified for {name}")
    return datasets
```

```python
datasets = {
    "accidents": df_accidents_new,
    "refinery_capacity": df_refinery_capacity,
    "refinery_inputs": df_refinery_inputs,
    "crude_rigs": df_crude_rigs,
    "refinery_utilization": df_refinery_utilization
}

columns = {
    "accidents": ["LOCAL_DATETIME"],
    "refinery_capacity": ["Date"],
    "refinery_inputs": ["Date"],
    "crude_rigs": ["Date"],
    "refinery_utilization": ["Date"]
}

datasets = extract_date_parts(datasets, columns)
```

```python
#Dropping the IYEAR column from the accidents data
df_accidents_new = df_accidents_new.drop(columns=["IYEAR"])
```

```python
#Saving the new datasets
import os

data_path = os.path.join("..", "data", "raw")
os.makedirs(data_path, exist_ok=True)

datasets = {
    "accidents": df_accidents_new,
    "refinery_capacity": df_refinery_capacity,
    "refinery_inputs": df_refinery_inputs,
    "crude_rigs": df_crude_rigs,
    "refinery_utilization": df_refinery_utilization
}

for name, df in datasets.items():
    save_path = os.path.join(data_path, f"{name}.csv")
    df.to_csv(save_path, index=False)
    print(f"Saved {name}.csv to {save_path}")
print("All datasets saved")
```
