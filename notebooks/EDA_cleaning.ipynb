```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

accidents = pd.read_csv("/Users/mac/Documents/MySchoolDocs/PERSONAL/JOB APPLICATIONS/ENERGY_PROJECT/data/raw/accidents.csv")
crude_rigs = pd.read_csv("/Users/mac/Documents/MySchoolDocs/PERSONAL/JOB APPLICATIONS/ENERGY_PROJECT/data/raw/crude_rigs.csv")
refinery_capacity = pd.read_csv("/Users/mac/Documents/MySchoolDocs/PERSONAL/JOB APPLICATIONS/ENERGY_PROJECT/data/raw/refinery_capacity.csv")
refinery_inputs = pd.read_csv("/Users/mac/Documents/MySchoolDocs/PERSONAL/JOB APPLICATIONS/ENERGY_PROJECT/data/raw/refinery_inputs.csv")
refinery_utilization = pd.read_csv("/Users/mac/Documents/MySchoolDocs/PERSONAL/JOB APPLICATIONS/ENERGY_PROJECT/data/raw/refinery_utilization.csv")
```

```python
def describe_data(datasets):
    for name, df in datasets.items():
        print(f"{name}")
        display(df.describe(include = "all"))
```

```python
datasets = {
    "accidents": accidents,
    "refinery_capacity": refinery_capacity,
    "refinery_inputs": refinery_inputs,
    "crude_rigs": crude_rigs,
    "refinery_utilization": refinery_utilization
}

describe_data(datasets)
```

```python
#Evaluating the missingness in the SOIL_CONTAMINATION

accidents["SOIL_CONTAMINATION"].value_counts(dropna=False)
```

```python
accidents.groupby("ONSHORE_STATE_ABBREVIATION")["SOIL_CONTAMINATION"].apply(lambda x: x.isna().sum()).sort_values(ascending=False)
```

```python
#Checking the soil_contamination missing values by state

accidents.groupby("ONSHORE_STATE_ABBREVIATION")["SOIL_CONTAMINATION"].apply(lambda x: x.isna().mean()*100).sort_values(ascending=False)
```

Because missingness is non-random and geographically concentrated, filling with "YES" (the mode) will distort the regional contamination patterns and overstate the contamination levels in some states that have incomplete reporting. As such, we fill all missing SOIL_CONTAMINATION values with "Unknown".

```python
accidents[accidents['ONSHORE_STATE_ABBREVIATION'].isna()]['LOCAL_DATETIME_Year'].value_counts(dropna=False)
```

Since the missing values for ONSHORE_STATE_ABBREVIATION are spread across many years and not concentrated in one period, we fill it with the mode (TX).

```python
#Filling all na for SOIL_CONTAMINATION with "Unknown"

accidents["SOIL_CONTAMINATION"].fillna("Unknown", inplace=True)

#Filling all missing state abbreviations with mode (TX)
accidents["ONSHORE_STATE_ABBREVIATION"].fillna(accidents["ONSHORE_STATE_ABBREVIATION"].mode()[0], inplace=True)

#Filling all missing values of EST_COST_PROP_DAMAGE and EST_COST_ENVIRONMENTAL with their respective medians.
accidents["EST_COST_PROP_DAMAGE"].fillna(accidents["EST_COST_PROP_DAMAGE"].median(), inplace=True)
accidents["EST_COST_ENVIRONMENTAL"].fillna(accidents["EST_COST_ENVIRONMENTAL"].median(), inplace=True)
```

```python
#Renaming columns in crude_rigs, refinery_capacity, refinery_inputs and refinery_utilization datasets

crude_rigs.rename(columns={"U.S. Crude Oil Rotary Rigs in Operation (Count)":"Operational_Rigs"}, inplace=True)
refinery_capacity.rename(columns={"U. S. Operable Crude Oil Distillation Capacity  (Thousand Barrels per Calendar Day)": "Distillation_Capacity"}, inplace=True)
refinery_inputs.rename(columns={"U.S. Gross Inputs to Refineries (Thousand Barrels Per Day)": "Refinery_Inputs"}, inplace=True)
refinery_utilization.rename(columns={"U.S. Percent Utilization of Refinery Operable Capacity":"%_Utilization_of_Refinery_Operable_Capacity"}, inplace=True)
```

## Data Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Yearly average Operational Rig counts from 1973 to 2025

yearly_crude_rigs = (crude_rigs.groupby("Date_Year", as_index=False)["Operational_Rigs"].mean())

plt.figure(figsize=(10,6))
plt.plot(yearly_crude_rigs["Date_Year"],yearly_crude_rigs["Operational_Rigs"], marker="o")
plt.title("Yearly Average Operation Rig Counts (1973-2025)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#Monthly average operational rig counts from 1973-2025

monthly_crude_rigs = (crude_rigs.groupby("Date_Month", as_index=False)["Operational_Rigs"].mean())

plt.figure(figsize=(10,6))
plt.bar(monthly_crude_rigs["Date_Month"], monthly_crude_rigs["Operational_Rigs"])
plt.title("Monthly Average Operational Rigs from 1973-2025")
plt.xlabel("Month")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
